```python
def income_data(sal):
    if sal== '<=50K':
        return 0
    else:
        return 1
        
```

```python
df['encoded_income']= df['income'].apply(income_data)
```

```python
df.head(1)
```

```python
df.replace(to_replace=['<=50K','>=50K'],value=[0,1],inplace = True)
```

```python
#  which workclass getting highest salary
       
df.groupby('workclass')['encoded_income'].mean().sort_values(ascending = False)
```

```python
#  How has better chance to get salary >50k male or female
df.groupby('gender')['encoded_income'].mean().sort_values(ascending=False)
```

```python
# Convert workclass column datatype to category datatype
df['workclass']=df['workclass'].astype('category')
```
